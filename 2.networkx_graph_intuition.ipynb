```python
import networkx as nx 
```

```python
G = nx.Graph()           # create graph

G.add_node(1)              # nodes(or vertices) add node       
G.add_nodes_from([2,3])    # Here add multiple nodes


G.add_edge(1,2)            # add edges
e=(2,3)
G.add_edge(*e)                     # * unpacks tuple and then add() adds edges
G.add_edges_from([(1,2), (1,3)])   # add multiple edges
```

```python
G.edges()         # edges(or links)   show all edges
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

nx.draw_networkx(G)     # to visualize graph 
plt.axis('off')
plt.draw()
```

## Modelling  road network of India's cities

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

G = nx.Graph()    # undireted graph , here we are working with this

# G = nx.DiGraph()   # directed graph

city_set=['Dehi', 'Bangalore', 'Pune', 'Mumbai', 'Chennai', 'Calcutta', 'Jammu', 'Surat', 'Patna', 'Jaipur']  # used list

# 1.Add nodes
for city in city_set:
    G.add_node(city)              
    
# 2.Add weights for edges    
costs=[]
value=100                 # used to add weights for edges when we add to nodes
while (value<=2000):
    costs.append(value)   # it append like 100, 200, 300,.....,2000
    value += 100
# print(costs)    

    
# 3.Add edges , adding 20 edges to this network
while(G.number_of_edges() < 20):        # number_of_edges() runs upto 20 to add edges
    
    # random.choice() picks random city from nodes for -> c1 , also pass list(G.nodes)
    c1 = random.choice(list(G.nodes()))       
    c2 = random.choice(list(G.nodes()))      # similar for c2
    
    # for creating edges 2 condition : c1, c2 cities not be equal  and  
    # there should not be an edge which already present b/w these 2 nodes
    # has_edge() : tells whether there is an edge b/w 2 nodes should be equal to 0
    if c1 != c2 and G.has_edge(c1,c2)==0:     
        w = random.choice(costs)              # adding weights
        G.add_edge(c1, c2, weight=w)          # finally add edge and weight
    
    
pos = nx.circular_layout(G)         # layouts for graph like : nx.spring_layout() , nx.spectral_layout()
nx.draw(G, pos, with_labels=True)    # pass parameter - pos also

nx.draw_networkx_edge_labels(G, pos)     # it will assign weights to edges

plt.show()
```

```python
# it tells graph is connected , every 2 cities nodes are connected

nx.is_connected(G)        

# if we reduce the nmber of edges (G.number_of_edges < 10) like to 10 we created above, it lets some nodes not connected and 
# then we check is_connected()  , it gives --> False.
```

```python
# gives path of  every node connected  or not

for u in G.nodes():
    for v in G.nodes():
        print(u,v,nx.has_path(G,u,v))
```
